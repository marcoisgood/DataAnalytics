```python
ZooData['waterstronge'] = (ZooData.aquatic + ZooData.predator + ZooData.toothed + ZooData.backbone)
ZooData['landstronge'] = (ZooData.legs + ZooData.predator + ZooData.toothed + ZooData.backbone)
ZooData['rate'] = round((ZooData.landstronge+1)/(ZooData.waterstronge+1))
ZooData['rate']=ZooData['rate'].astype(int)
ZooData
```

## kNN function

```python
import numpy as np
from numpy import
import matplotlib.pyplot as plt
from sklearn import datasets

newZoo = ZooData[["animal_name","feathers","eggs","milk"]]
x = newZoo.feathers
y = newZoo.eggs

plt.scatter(x[y==0], x[y==1], color = 'red')
plt.scatter(x[y==1], x[y==0], color = 'green')
plt.scatter(0.8,0.5, color = 'blue')
x_1 = np.array(0.8,0.5)


#
distances=[np.sqrt(np.sum((x_t-x_1)**2)) for x_t in x]
#
d=np.sort(distances)
nearest=np.argsort(distances)
k=4
topk_y=[y[i] for i in nearest[:k]]
from collections import Counter
#topk_y
votes=Counter(topk_y)
#1
print(votes)
predict_y=votes.most_common(1)[0][0]
print(predict_y)
plt.show()
```

## Your friendly neighborhood, SpiderMan

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, p = 2)# p=2 for euclidean distance
knn.fit(ZooData[["waterstronge", "landstronge"]], ZooData.rate)

predictionData = ZooData[['waterstronge','landstronge','rate']]
predictionData['rate'].value_counts()
predictionData.head()

fig = plt.figure(figsize=(10,10))
colormap = np.array(['black', 'yellow', 'red', 'blue', 'green','purple','white'])
 
plt.scatter(predictionData.waterstronge, predictionData.landstronge, 
            c = colormap[ZooData.rate], s = 120)
plt.title('type:\n 1 - Yellow\n 2 - Red\n 3 - Blue\n 4 - green\n')
ax = fig.add_subplot(111)
ax.set_xlabel('strong')
ax.set_ylabel('care')
```

## Prediction Mesh

```python
def plotMesh():
    h = 100 
    # Create color maps
    cmap_light = ListedColormap(['#ffffb3', '#ff9999','#d6d6f5','#ccffdd','#A52A2A','#9932CC'])
    colormap = np.array(['black', 'yellow', 'red', 'blue', 'green','brown','purple'])

    x_min, x_max = ZooData.waterstronge.min(), ZooData.waterstronge.max()
    y_min, y_max = ZooData.landstronge.min(), ZooData.landstronge.max() 
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    fig = plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap = cmap_light)

    # Plot also the training points
    plt.scatter(ZooData.waterstronge, ZooData.landstronge, c = colormap[ZooData.class_type], 
                s = 120)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("4-Class classification \n(k = %i)\n\
    Loan 1 - Yellow, Loan 2 - Red, Loan 3 - Blue, Loan 4 - green" % (5))
    ax = fig.add_subplot(111)
    ax.set_xlabel('Water Predator')
    ax.set_ylabel('Land Predator')
plotMesh()
```
